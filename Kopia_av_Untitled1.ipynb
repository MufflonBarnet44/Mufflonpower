````python
# Import the required libraris
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
````

# Step2: Load the FashinMNIST Dataset and Slit Into Training and Validation **Sets**

````python
# Create a transaform to preprocess the image
# This step converts an image into a multi-dimentianal array of data

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load the FashinMNIST datastet

train_data = datasets.FashionMNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

# Split DAta into traning and validation set

num_train = len(train_data)
num_val = int(0.1 * num_train)
train_data, val_data = random_split(train_data, [num_train - num_val, num_val])

# Define the Data loader
batch_size = 32 # You can adjust this value as needed

train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)

print("Data loaded and Dataloaders are ready")



````
